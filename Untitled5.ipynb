```python
import pandas as pd
data = pd.read_csv("/content/porsche_911.csv")
#поиск дубликатов:
data.duplicated(keep='first')
```

удаление дубликатов

```python
data.drop_duplicates(keep='first')
```

разделить таблицу на две новых, и объединить их разными способами

```python
df1=data[['link','maximum_speed','power_per_litre']]
df2=data[['link','engine','start_of_production']]
result1 = pd.merge(df1,df2,on='link',how='outer')
result2 = pd.merge(df1,df2,on='link',how='inner')
result3 = pd.merge(df1,df2,on='link',how='left')
result4 = pd.merge(df1,df2,on='link',how='right')
result1
```

выполнить нормализацию численных данных

```python
from sklearn.preprocessing import normalize
df_n= data[['maximum_speed','power_per_litre']]
df_n1 = df_n.fillna(value=0)
normalized_data = normalize(df_n1, norm='l1', axis=1)
normalized_data
```
